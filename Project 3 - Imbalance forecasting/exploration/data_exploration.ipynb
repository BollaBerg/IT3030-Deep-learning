# Analyse target columns
The second assignment is to plot an altered forecasting, looking at structural inbalance (the difference between slowly changing consumption and the abruptly changing power production), instead of the total grid imbalance. To do this, we must first find a good interpolation method for the target.

```python
plt.plot(train_data["y"])
```

```python
# Look at a single week
elements_to_see = int((60 / 5) * 24 * 4)
plt.plot(train_data["y"][0:elements_to_see])
```

```python
window = 12 * 6
start_watching = 12 * 24 * 60
averages = train_data["y"].rolling(window, center=True).mean()
interpolation = np.interp(train_data["y"].index, averages.index, averages)
interpolation[:int(window/2)] = np.mean(train_data["y"][:int(window/2)])
interpolation[-(int(window/2) - 1):] = np.mean(train_data["y"][-(int(window/2) - 1):])

plt.plot(range(elements_to_see), train_data["y"][start_watching:start_watching + elements_to_see])
plt.plot(interpolation[start_watching:start_watching + elements_to_see])
```

```python
plt.plot(train_data["y"] - interpolation)
```
